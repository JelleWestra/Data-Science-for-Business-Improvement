```python
# -*- coding: utf-8 -*-
"""
Created on Sat Dec 12 12:45:25 2020

@author: Jelle
"""

import pygame
import sys
import numpy as np

FRAME_RATE = 60

def to_unit(v):
    try:
        return v / np.linalg.norm(v, axis=1).reshape(-1, 1)
    except:
        return v / np.linalg.norm(v)
    

class Box():
    def __init__(self, n_people=10, infection_rate=0.2):
        self.n_people = n_people
        self.infection_rate = infection_rate
        self.dt = 1/FRAME_RATE
        
        self.size = 256
        self.p = np.random.uniform(-self.size/2, self.size/2, (self.n_people, 2))
        
        self.state = np.zeros(self.p.shape)
        self.I = 1
        self.radius = 10
        for _ in range(self.I):
            i = np.random.randint(self.p.shape[0])
            self.state[i,0] = 1 
        
        self.v_max = 50
        v_dir = np.random.uniform(-1, 1, self.p.shape)
        self.v = self.v_max * v_dir / np.linalg.norm(v_dir)
                
        self.k = 500
        
        self.F = np.zeros(self.p.shape)
        self.F_rand = self.k / 10 * to_unit(np.random.uniform(-1, 1, self.F.shape))
        
        
    def update(self, frame):
        if frame%FRAME_RATE == 0:
            self.state[self.state[:,0] == 1, 1] += 1
        
        for i in range(self.p.shape[0]):
            if self.state[i, 0] == 1:
                if self.state[i, 1] >= 4:
                    self.state[i, 0] = 2
                else:
                    idx = (np.linalg.norm(self.p - self.p[i], axis=1) < self.radius) & (self.state[:,0] < 1)
                    self.state[:,0] += idx * np.random.choice([0, 1], size=self.p.shape[0], p=[1-self.infection_rate, self.infection_rate])
        
        
        self.F = -self.k * np.sign(self.p) * (np.abs(self.p) - .6 * self.size/2)/(.4 * self.size/2) * (np.abs(self.p) > .6 * self.size/2)

        if frame%FRAME_RATE == 0:
            self.F_rand = self.k / 10 * to_unit(np.random.uniform(-1, 1, self.F.shape))
        
        self.F += self.F_rand
   
        self.v += self.F * self.dt
        self.v *= self.v_max / np.linalg.norm(self.v, axis=1).reshape(-1, 1)
        self.p += self.v * self.dt
        
        
        
b = Box(100, 0.05)

# General setup
pygame.init()
clock = pygame.time.Clock()

# Setting up main window
screen_width = 512
screen_height = 512
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Corono simulator 2020')

# Game rectangles
person_size = 5
persons = [pygame.Rect(0,0, person_size, person_size) for p in b.p]

bg_color = pygame.Color('black')
white = pygame.Color('white')
red = pygame.Color('indianred1')
dark_gray = pygame.Color('gray48')

font = pygame.font.SysFont('Georgia', 18)
 
def update_fps():
    fps = str(int(clock.get_fps())) + ' fps'
    fps_text = font.render(fps, 1, white)
    return fps_text

def display_T(frame):
    T = f'T = {frame/FRAME_RATE:.1f}'
    return font.render(T, 1, white)

STATES_HISTORY = []

frame = 0
while True:
    # Handling input
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Visuals
    screen.fill(bg_color)
    
    screen.blit(update_fps(), (10,0))
    T_txt = display_T(frame)
    screen.blit(T_txt, (screen_width/2 - T_txt.get_rect().width/2, screen_height/2 - b.size/2 - T_txt.get_rect().height - 10))
    
    for i, p in enumerate(persons):
        if b.state[i,0] == 0:
            pygame.draw.ellipse(screen, white, p)
        elif b.state[i,0] == 1:
            pygame.draw.ellipse(screen, red, p)
        else:
            pygame.draw.ellipse(screen, dark_gray, p)
        
        p.x, p.y = b.p[i] + np.array([screen_width/2 - 5, screen_height/2 - 5])
    
    
    state = [0,0,0]
    for s in b.state[:,0]:
        state[int(s)] += 1
    STATES_HISTORY.append(state)
    
    b.update(frame)
        
    # draw a rectangle
    pygame.draw.rect(screen, white, pygame.Rect(-b.size/2 + screen_width/2 - person_size/2, -b.size/2 + screen_height/2 - person_size/2, b.size, person_size/2))
    pygame.draw.rect(screen, white, pygame.Rect(b.size/2 + screen_width/2 - person_size/2, -b.size/2 + screen_height/2 - person_size/2, person_size/2, b.size))
    pygame.draw.rect(screen, white, pygame.Rect(-b.size/2 + screen_width/2 - person_size/2, b.size/2 + screen_height/2 - person_size/2, b.size + person_size/2, person_size/2))
    pygame.draw.rect(screen, white, pygame.Rect(-b.size/2 + screen_width/2 - person_size/2, -b.size/2 + screen_height/2 - person_size/2, person_size/2, b.size))
    
    # Updating the window
    pygame.display.flip()
    clock.tick(FRAME_RATE)
    frame += 1
```

```python
A = np.array(STATES_HISTORY)

import matplotlib.pyplot as plt

plt.plot(A)
plt.show()
```
